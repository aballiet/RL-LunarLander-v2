```python
import matplotlib.pyplot as plt
import csv

```

```python
with open("./models/DQN_simple/rewards.csv") as csvfile:
    DQN_simple = list(csv.reader(csvfile))

DQN_simple = list(map(float,DQN_simple[0]))
print(DQN_simple)
```

```python
with open("./models/DQN_double/rewards.csv") as csvfile:
    DQN_double = list(csv.reader(csvfile))

DQN_double = list(map(float,DQN_double[0]))
print(DQN_double)
```

```python
fig = plt.figure(1, figsize=(20, 10))

plt.plot(DQN_simple, label='Simple DQN')
plt.plot(DQN_double, label='Double DQN')

plt.legend(loc='lower left')
plt.xlabel('episodes', fontsize=20)
plt.ylabel('rewards', fontsize=20)

plt.ylim(-600, 300)

plt.show()
```
